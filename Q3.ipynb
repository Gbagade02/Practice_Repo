```python
import gdown
import pandas as pd
import matplotlib.pyplot as plt

file_id = '1F_LOzrBtu4EAbmb2bx9SufprVDXk0Atm'
url = f'https://drive.google.com/uc?id={file_id}'

output_file = 'stock_prices.csv'  # The name you want to save the downloaded file as

gdown.download(url, output_file)
```

```python
# Read the CSV file into a DataFrame
data = pd.read_csv('stock_prices.csv')

# Convert the 'date' column to a datetime object
data['date'] = pd.to_datetime(data['date'])

# Sort the DataFrame by date
data.sort_values(by='date', inplace=True)

# Calculate the 30-day rolling moving average
data['rolling_average'] = data['closing_price'].rolling(window=30).mean()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(data['date'], data['closing_price'], label='Closing Price')
plt.plot(data['date'], data['rolling_average'], label='30-day Rolling Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Prices and 30-day Rolling Average')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
